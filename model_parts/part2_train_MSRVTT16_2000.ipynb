```python
import pickle
import json
import os
import numpy as np
from keras.layers import TextVectorization
from keras.utils import to_categorical
from keras.models import Model
from keras.layers import Input, LSTM, Dense
from keras.optimizers import Adam
from keras.callbacks import Callback
from keras.callbacks import ReduceLROnPlateau
from keras.utils import plot_model
import matplotlib.pyplot as plt

# The parameters of the model:
batch_size = 256
learning_rate = 0.0002
epochs = 100
latent_dim = 512 # LSTM hidden node size
max_length = 10  # Number of tokens in one caption
vectorizer = None

num_enc_tokens = 4096 # Since VGG16 output is features with dimension 4096
num_dec_tokens = 2000 # Vocabulary size
time_steps_enc = 16 # vs 80 frames per video
time_steps_dec = 10

feats_path = "/content/drive/My Drive/train_MSRVTT16/train_features16/"
final_model_path = "/content/drive/My Drive/train_MSRVTT16/train_2000Tokens/models_MSRVTT16_2000/"
caps_path = "/content/drive/My Drive/train_MSRVTT16/train_val_videodatainfo.json"
ckpt_model_path = '/content/drive/My Drive/train_MSRVTT16/train_2000Tokens/checkpoints_2000/'

##### Saves models only with certain val_accuracy defined by threshold
class ThresholdCallback(Callback):
    def __init__(self, threshold1, enc_inputs, enc_states, dec_inputs):
        super(ThresholdCallback, self).__init__()
        self.threshold1 = threshold1
        #self.threshold2 = threshold2
        self.enc_inputs = enc_inputs
        self.enc_states = enc_states
        self.dec_inputs = dec_inputs

    def on_epoch_end(self, epoch, logs=None):
        val_acc = logs["val_accuracy"]
        if val_acc >= self.threshold1:
            enc_model = Model(self.enc_inputs, self.enc_states)
            enc_model.save(os.path.join(ckpt_model_path, 'enc_model' + str(epoch) + '.h5'))

            dec_state_input_h = Input(shape=(latent_dim,))
            dec_state_input_c = Input(shape=(latent_dim,))
            dec_states_inputs = [dec_state_input_h, dec_state_input_c]
            dec_outputs, state_h, state_c = dec_lstm(self.dec_inputs, initial_state=dec_states_inputs)
            dec_states = [state_h, state_c]
            dec_outputs = dec_dense(dec_outputs)
            dec_model = Model([self.dec_inputs] + dec_states_inputs, [dec_outputs] + dec_states)
            dec_model.save_weights(os.path.join(ckpt_model_path, 'dec_model_weights' + str(epoch) + '.h5'))

        # If training needs to be stopped:
        val_loss = logs["val_loss"]
        if np.isnan(val_loss):
            self.model.stop_training = True

#####Data generator for training
def gen_data(data_caps, X_data, vectorizer):

    vidCaps = []
    vidIds = []
    for capID in data_caps:
        vidCaps.append(capID[0]) # Taking only captions
        vidIds.append(capID[1]) # Taking only video ids

    caps_vector = vectorizer(vidCaps) # Tokenizing captions
    caps_vector = np.array(caps_vector)
    caps_size = len(caps_vector) # Number of captions

    count = 0 # Counter of batch size
    enc_input_data = []
    dec_input_data = []
    dec_output_data = []
    for i in range(epochs): # Every epochs
        for j in range(caps_size): # Every batch size
            enc_input_data.append(X_data[vidIds[j]]) # Obtaining features
            Y_data = to_categorical(caps_vector[j], num_dec_tokens) # Obtaining categorized and tokenized captions
            dec_input_data.append(Y_data[:-1]) # Taking without <eos>
            dec_output_data.append(Y_data[1:]) # Taking without <bos>
            count += 1 # Counting batch size
            if count == batch_size: # When it reaches the batch size
                count = 0 # Refreshing the counter of batch size
                enc_input = np.array(enc_input_data)
                dec_input = np.array(dec_input_data)
                dec_output = np.array(dec_output_data)
                enc_input_data = [] # Refreshing enc_input_data
                dec_input_data = [] # Refreshing enc_input_data
                dec_output_data = [] # Refreshing enc_input_data
                yield ([enc_input, dec_input], dec_output)

# Loading a json file with MSRVTT captions for all videos (7010)
# In MSRVTT caption dataset, there are multiple captions for each video:
with open(caps_path) as file:
    y_data = json.load(file) # contains captions and their video ids

# Obtaining video ids (names) of train and valid data:
train_vidIDs = []
valid_vidIDs = []
for i in range(0, len(y_data['videos'])):
    if y_data['videos'][i]['split'] == 'train':
        train_vidIDs.append(y_data['videos'][i]['video_id'])
    else:
        valid_vidIDs.append(y_data['videos'][i]['video_id'])
print("Size of train dataset:", len(train_vidIDs))
print("Size of validation dataset:", len(valid_vidIDs))

data_capID = [] # contains all the captions with their video ID

# Inserting tokens <bos> and <eos> at the beginning and ending of each caption respectively:
for i in range(0, len(y_data['sentences'])):
    cap = y_data['sentences'][i]['caption']
    vidID = y_data['sentences'][i]['video_id']
    cap = cap.split(" ")
    cap.insert(0, "<bos>")
    cap.append("<eos>")
    # Taking captions with length only between 6 and 13
    if len(cap) >= 6 and len(cap) <= 13:
        cap = ' '.join(cap)
        data_capID.append([cap, vidID])

# Dividing captions to train and valid as in caption dataset:
train_caps = []
valid_caps = []
for i in range(0, len(data_capID)):
    for vidID in train_vidIDs:
        if data_capID[i][1] == vidID:
            train_caps.append(data_capID[i])
    for vidID in valid_vidIDs:
        if data_capID[i][1] == vidID:
            valid_caps.append(data_capID[i])
print("Size of train captions:", len(train_caps))
print("Size of validation captions:", len(valid_caps))

# These are dictionaries which will contain train/valid video names as keys
# and extracted features as values (from 1st part: feature extraction):
x_data_train = {}
x_data_valid = {}
for filename in os.listdir(feats_path):
    feat = np.load(os.path.join(feats_path, filename), allow_pickle=True)
    if filename[:-8] in train_vidIDs:
        x_data_train[filename[:-8]] = feat
    else:
        x_data_valid[filename[:-8]] = feat

vocabulary = [] # vocabulary contains only captions from training data
for capID in train_caps:
    vocabulary.append(capID[0])

# vectorizer: every caption is padded and every word gets a unique integer value
# vectorizer's vocabulary is based only on captions of train dataset:
vectorizer = TextVectorization(
    max_tokens= num_dec_tokens,
    standardize=None,
    split="whitespace",
    output_mode="int",
    output_sequence_length=max_length,
)
vectorizer.adapt(vocabulary)

# Now, let's prepare for training by getting data from data generator
train_data = gen_data(train_caps, x_data_train, vectorizer)
valid_data = gen_data(valid_caps, x_data_valid, vectorizer)
steps_per_epoch = len(train_caps)//batch_size # Number of batches in train_data per epoch
validation_steps = len(valid_caps)//batch_size # Number of batches in valid_data per epoch

# Designing model:
enc_inputs = Input(shape=(time_steps_enc, num_enc_tokens), name="enc_inputs")
enc_lstm = LSTM(latent_dim, return_state=True, return_sequences=True, name='enc_lstm')
_, state_h, state_c = enc_lstm(enc_inputs)
enc_states = [state_h, state_c]

dec_inputs = Input(shape=(time_steps_dec, num_dec_tokens), name="dec_inputs")
dec_lstm = LSTM(latent_dim, return_sequences=True, return_state=True, name='dec_lstm')
dec_outputs, _, _ = dec_lstm(dec_inputs, initial_state=enc_states)
dec_dense = Dense(num_dec_tokens, activation='relu', name='dec_relu')
dec_outputs = dec_dense(dec_outputs)

model = Model([enc_inputs, dec_inputs], dec_outputs)
print("Overall Model summary")
model.summary()
plot_model(model, "/content/drive/My Drive/train_MSRVTT16/train_2000Tokens/model_16_2000.png", show_shapes=True, show_dtype=True, show_layer_names=True, rankdir='BT', show_layer_activations=True)

# Optimizer:
optimizer = Adam(learning_rate=learning_rate)

# Callbacks:
my_callback = ThresholdCallback(threshold1=0.4, enc_inputs=enc_inputs, enc_states=enc_states, dec_inputs=dec_inputs) # threshold1 = val_accuracy
reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=4)

model.compile(metrics=['accuracy'], optimizer=optimizer, loss='categorical_crossentropy')
history = model.fit(train_data, validation_data=valid_data, epochs=epochs, steps_per_epoch=steps_per_epoch, validation_steps=validation_steps, callbacks=[my_callback, reduce_lr])

print("The list of all data in history:", history.history.keys())
# Plotting accuracy vs epoch
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
# Plotting loss vs epoch
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()
# Plotting lr vs epoch
plt.plot(history.history['lr'])
plt.title('Model Learning Rate')
plt.ylabel('Learning Rate')
plt.xlabel('Epoch')
plt.show()

# Saving Vectorizer:
with open(os.path.join(final_model_path,'vectorizer' + str(num_dec_tokens) + ".pickle"), "wb") as file:
    save_vectorizer = {"vec_config":  vectorizer.get_config(), "vec_weights": vectorizer.get_weights()}
    pickle.dump(save_vectorizer, file)

# Saving models:
# This is the model saved on the last epoch
enc_model = Model(enc_inputs, enc_states)
enc_model.save(os.path.join(final_model_path, 'enc_model.h5'))
print("Encoder Model summary")
enc_model.summary()
plot_model(enc_model, "/content/drive/My Drive/train_MSRVTT16/train_2000Tokens/enc_model_16_2000.png", show_shapes=True, show_dtype=True, show_layer_names=True, rankdir='BT', show_layer_activations=True)

dec_state_input_h = Input(shape=(latent_dim,))
dec_state_input_c = Input(shape=(latent_dim,))
dec_states_inputs = [dec_state_input_h, dec_state_input_c]
dec_outputs, state_h, state_c = dec_lstm(dec_inputs, initial_state=dec_states_inputs)
dec_states = [state_h, state_c]
dec_outputs = dec_dense(dec_outputs)
dec_model = Model([dec_inputs] + dec_states_inputs, [dec_outputs] + dec_states)
print("Decoder Model summary")
dec_model.summary()
plot_model(dec_model, "/content/drive/My Drive/train_MSRVTT16/train_2000Tokens/dec_model_16_2000.png", show_shapes=True, show_dtype=True, show_layer_names=True, rankdir='BT', show_layer_activations=True)
dec_model.save_weights(os.path.join(final_model_path, 'dec_model_weights.h5'))
```
