```python
import pickle
import json
import os

# path to the MSVD caption dataset:
MSVD_caps_path = "raw-captions.pkl"
# path to the MSRVTT caption dataset:
MSRVTT_caps_path  = "test_videodatainfo.json"
# path to the MSVD test features dataset:
MSVD_test_path = "MSVD_FEATURES/feats_MSVD16/Test_feat/"
# path to the MSRVTT test features dataset:
MSRVTT_test_path = "MSRVTT_FEATURES/feats_MSRVTT16/test_features16/"

# Creating a json file of MSVD test captions:
with open(MSVD_caps_path, 'rb') as file:
    MSVD_caps = pickle.load(file)

MSVD_list = []
for filename in os.listdir(MSVD_test_path):
    temp_dict1 = {}
    temp_dict1['video_id'] = filename[:-8]
    temp_dict1['captions'] = []
    for cap in MSVD_caps[filename[:-8]]:
        new_cap = ' '.join(cap)
        temp_dict1['captions'].append([new_cap])
    MSVD_list.append(temp_dict1)

MSVD_json = json.dumps(MSVD_list, indent = 4)

with open("MSVD_test_captions.json", "w") as outfile:
    outfile.write(MSVD_json)

# Creating a json file of MSRVTT test captions:
with open(MSRVTT_caps_path) as file:
    MSRVTT_caps = json.load(file)

MSRVTT_list = []
for filename in os.listdir(MSRVTT_test_path):
    temp_dict2 = {}
    temp_dict2['video_id'] = filename[:-8]
    temp_dict2['captions'] = []
    for i in range(0, len(MSRVTT_caps['sentences'])):
        if MSRVTT_caps['sentences'][i]['video_id'] == filename[:-8]:
            caption = MSRVTT_caps['sentences'][i]['caption']
            temp_dict2['captions'].append([caption])
    MSRVTT_list.append(temp_dict2)

MSRVTT_json = json.dumps(MSRVTT_list, indent = 4)

with open("MSRVTT_test_captions.json", "w") as outfile:
    outfile.write(MSRVTT_json)
```
